```python
import gym
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque
import matplotlib.pyplot as plt
from torch.utils.tensorboard import SummaryWriter
from datetime import datetime
import os
from tqdm import tqdm

# ---------------------
# Load Data
# ---------------------
df = pd.read_parquet('./Data/combined_dataset.parquet')
df = df.fillna(0)  # Clean any missing data
df = df.sort_values('timestamp')  # Ensure time order

# Define columns for normalization
features = ['active_power', 'battery_soc']
df[features] = (df[features] - df[features].mean()) / df[features].std()

# ---------------------
# Train/Test Split
# ---------------------
train_size = int(len(df) * 0.8)
train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:].reset_index(drop=True)

# ---------------------
# Environment with Simulated Battery
# ---------------------
class EnergyEnv(gym.Env):
    def __init__(self, dataframe):
        super(EnergyEnv, self).__init__()
        self.df = dataframe.reset_index(drop=True)
        self.current_step = 0
        self.battery_capacity = 100.0  # Example capacity
        self.battery_level = 50.0      # Start at 50%
        self.max_steps = len(self.df) - 1
        self.action_space = gym.spaces.Discrete(3)  # 0: Charge, 1: Discharge, 2: Idle
        # Observation: active_power and normalized battery level (2 features)
        self.observation_space = gym.spaces.Box(low=-np.inf, high=np.inf, shape=(2,), dtype=np.float32)

    def reset(self):
        self.current_step = 0
        self.battery_level = 50.0
        return self._get_obs()

    def _get_obs(self):
        row = self.df.iloc[self.current_step]
        return np.array([row['active_power'], self.battery_level / self.battery_capacity], dtype=np.float32)

    def step(self, action):
        done = False
        row = self.df.iloc[self.current_step]
        inverter_power = row['active_power']  # Using active_power
        # Use a scaling factor to reduce reward magnitudes
        scale = 0.001
        reward = 0

        if action == 0:  # Charge
            self.battery_level = min(self.battery_level + 1, self.battery_capacity)
            reward = scale * (-abs(inverter_power))
        elif action == 1:  # Discharge
            if self.battery_level > 0:
                self.battery_level = max(self.battery_level - 1, 0)
                reward = scale * (-abs(inverter_power))
            else:
                reward = -1  # Reduced penalty for discharging empty battery
        elif action == 2:  # Idle
            reward = scale * (-abs(inverter_power))

        self.current_step += 1
        if self.current_step >= self.max_steps:
            done = True

        return self._get_obs(), reward, done, {}

# ---------------------
# DQN Model
# ---------------------
class DQN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(DQN, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, output_dim)
        )

    def forward(self, x):
        return self.fc(x)

# ---------------------
# Agent Logic
# ---------------------
class DQNAgent:
    def __init__(self, state_size, action_size, writer=None):
        self.state_size = state_size
        self.action_size = action_size
        self.writer = writer
        self.model = self._build_model().to(device)
        self.target_model = self._build_model().to(device)
        self.target_model.load_state_dict(self.model.state_dict())
        self.memory = deque(maxlen=2000)
        self.gamma = 0.95  # Discount factor
        self.epsilon = 1.0  # Exploration rate
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.batch_size = 32
        self.learning_rate = 0.001
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.best_reward = float('-inf')

    def _build_model(self):
        model = nn.Sequential(
            nn.Linear(self.state_size, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, self.action_size)
        )
        return model

    def update_target_model(self):
        self.target_model.load_state_dict(self.model.state_dict())

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        state = torch.FloatTensor(state).unsqueeze(0).to(device)
        q_values = self.model(state)
        return torch.argmax(q_values, dim=1).item()

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def replay(self, episode):
        if len(self.memory) < self.batch_size:
            return

        batch = random.sample(self.memory, self.batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)
        states = torch.FloatTensor(np.array(states)).to(device)
        next_states = torch.FloatTensor(np.array(next_states)).to(device)
        actions = torch.LongTensor(np.array(actions)).to(device)
        rewards = torch.FloatTensor(np.array(rewards)).to(device)
        dones = torch.BoolTensor(np.array(dones)).to(device)

        q_values = self.model(states)
        next_q_values = self.target_model(next_states)

        target_q_values = q_values.gather(1, actions.unsqueeze(1)).squeeze(1)
        target = rewards + (1 - dones.float()) * self.gamma * torch.max(next_q_values, dim=1)[0]

        loss = nn.MSELoss()(target_q_values, target.detach())
        self.optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
        self.optimizer.step()

        if self.writer:
            self.writer.add_scalar('Loss/train', loss.item(), episode)

    def update_target(self):
        self.target_model.load_state_dict(self.model.state_dict())

# ---------------------
# Training Loop with Checkpointing & Real-Time Plotting
# ---------------------
def train_agent(env, agent, episodes=50, checkpoint_path='best_dqn.pth'):
    rewards_log = []
    plt.figure(figsize=(10, 5))
    plt.xlabel("Episode")
    plt.ylabel("Total Reward")
    plt.title("Reward over Episodes")
    plt.grid(True)
    plt.ion()  # Enable interactive mode for live plotting

    for e in tqdm(range(episodes), desc="Training Episodes"):
        state = env.reset()
        total_reward = 0
        done = False

        while not done:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            agent.replay(e)
            state = next_state
            total_reward += reward

        agent.update_target_model()
        # Update epsilon once per episode
        agent.epsilon = max(agent.epsilon_min, agent.epsilon * agent.epsilon_decay)
        rewards_log.append((e, total_reward, agent.epsilon))
        if agent.writer:
            agent.writer.add_scalar("Reward/train", total_reward, e)

        # Update the live plot
        plt.clf()
        episodes_arr = np.array([log[0] for log in rewards_log])
        rewards_arr = np.array([log[1] for log in rewards_log])
        plt.plot(episodes_arr, rewards_arr, marker="o", linestyle="-")
        plt.xlabel("Episode")
        plt.ylabel("Total Reward")
        plt.title("Reward over Episodes")
        plt.grid(True)
        plt.pause(0.01)

        print(f"Episode {e+1}/{episodes}, Total Reward: {total_reward:.2f}, Epsilon: {agent.epsilon:.3f}")

        if total_reward > agent.best_reward:
            agent.best_reward = total_reward
            torch.save(agent.model.state_dict(), checkpoint_path)

    plt.ioff()  # Disable interactive mode
    return rewards_log

# ---------------------
# Device Setup
# ---------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#print("Using device:", device)

# ---------------------
# TensorBoard Writer Setup
# ---------------------
logdir = f"runs/energy_rl_{datetime.now().strftime('%Y%m%d-%H%M%S')}"
writer = SummaryWriter(logdir)

# ---------------------
# Initialize and Train
# ---------------------
train_env = EnergyEnv(train_df)
agent = DQNAgent(train_env.observation_space.shape[0], train_env.action_space.n, writer)

print("Starting training...")
reward_log = train_agent(train_env, agent, episodes=50, checkpoint_path="best_dqn.pth")

# Save reward log to CSV
reward_df = pd.DataFrame(reward_log, columns=["episode", "reward", "epsilon"])
reward_df.to_csv("reward_log.csv", index=False)

# ---------------------
# Inference on Test Data
# ---------------------
test_env = EnergyEnv(test_df)
agent.model.load_state_dict(torch.load("best_dqn.pth"))
state = test_env.reset()
done = False
test_rewards = []

while not done:
    action = agent.act(state)
    state, reward, done, _ = test_env.step(action)
    test_rewards.append(reward)

print(f"Test Run - Total Reward: {sum(test_rewards):.2f}")

# ---------------------
# Final Reward Plot (saved)
# ---------------------
plt.figure(figsize=(10, 5))
plt.plot(reward_df["episode"], reward_df["reward"], marker="o")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("Reward over Episodes")
plt.grid(True)
plt.savefig("reward_plot.png")
plt.show()

writer.close()
```

# Reinforcement Learning for Energy Management: Insights, Findings, and Recommendations

## Project Overview

In this project, we developed and trained a Deep Q-Network (DQN) reinforcement learning (RL) agent to manage energy storage (battery) in a simulated environment using historical energy data. The goal was to use RL to optimize the decision-making process related to charging, discharging, or idling the battery based on real-time inverter power data.

We used the following framework and libraries for implementation:

- Gym: For creating the environment.

- PyTorch: For building the neural network and optimization.

- Pandas: For data manipulation and preprocessing.

- Matplotlib: For plotting rewards and performance.

## Findings and Insights

**Training Process**

The training of the RL agent was carried out using the data provided in the combined_dataset.parquet, which included features like active_power and battery_soc. The agent was tasked with selecting actions (charging, discharging, or idling) at each step, based on the state of the environment (current battery level and inverter power).

- Training Duration: The training process took approximately 16 hours to complete 50 episodes, with a total reward for the final episode of -2458.72.

- Final Reward: The agent performed poorly on the test set with a total reward of -145.19.

**Environment and Reward Structure**

The reward structure was designed to provide a penalty for discharging the battery when it was empty and a small reward or penalty for charging and idling. However, the reward scale used (scale = 0.001) seemed too small, possibly hindering the agent's ability to learn effectively.

Key points from the reward structure:

- Charge Action: A positive reward proportional to inverter power would be beneficial.

- Discharge Action: Discharge should have a more significant penalty, especially when the battery is empty.

- Idle Action: Idle actions were lightly penalized to encourage efficient energy usage.

**Model Architecture**

The model utilized a feedforward neural network (FNN) with two hidden layers of 64 neurons each. Despite this, the agent struggled to optimize its actions effectively, likely due to the model's limited capacity to capture complex patterns in the data.

- The current architecture may not have been complex enough for the task at hand.

- Adding more layers or increasing the number of neurons could improve the model’s ability to represent the environment better.

**Hyperparameters**

Several hyperparameters were used in the training process:

- Gamma (Discount Factor): Set to 0.95, indicating that future rewards were slightly less important than immediate ones.

- Learning Rate: Set to 0.001 for the Adam optimizer.

However, the results indicate that the chosen values were not optimal:

- The learning rate might be too high, causing unstable updates to the model.

- A slower epsilon decay or the use of epsilon annealing could allow the agent to explore more before converging to a greedy policy.

**Training Challenges and Performance**

- The model failed to achieve positive rewards on both training and test data. This suggests that the agent wasn’t learning an optimal policy.

- The training reward graph, while tracking some improvement, shows that the agent struggled to break out of suboptimal behaviors throughout training.

- Possible Reasons for Poor Performance:

Inadequate reward function design.

Insufficient training episodes (50 episodes might not be enough for convergence).

Exploration-exploitation balance may not have been optimized.

Inadequate model capacity for the task at hand.

**Pros**

- Clear Framework: Using Gym, PyTorch, and other libraries allowed for a clear structure of the project with well-defined components (environment, model, agent).

- Visualization with TensorBoard: The use of TensorBoard for real-time monitoring of the agent's progress was helpful for observing training dynamics and debugging.

- Reward Structure: The reward function, despite needing improvements, was set up to encourage both energy efficiency and proper battery management.

- Modular Design: The project was designed modularly, with separate components for the environment, agent, and model, making it easy to extend or modify.

**Cons**

- Suboptimal Model Performance: Despite long training durations, the agent failed to achieve significant positive rewards, highlighting that the model didn’t learn effectively.

- Reward Scaling: The reward scaling was too small, which could have hindered learning. More significant, meaningful rewards/penalties are essential for faster convergence.

- Limited Model Complexity: The neural network architecture might not have had enough capacity to capture the complexities of the task.

- Training Duration: Training for only 50 episodes is relatively short for deep reinforcement learning tasks, which may have contributed to underfitting.

**Recommendations for Future Work**

- Reward Function Improvement:

Increase the reward magnitude, especially for positive actions like charging with available energy.

Penalize idle and suboptimal discharge actions more severely.

Encourage the agent to maintain the battery at an optimal level (e.g., 40%-60%).

- Increase Training Duration:

Increase the number of training episodes (e.g., 200–500 episodes) to allow more time for the agent to explore and learn the environment.

Implement periodic evaluation on the test set to monitor progress and adjust hyperparameters as needed.

- Hyperparameter Tuning:

Experiment with different values for learning rate, gamma, and batch size.

Implement grid search or random search to optimize hyperparameters.

- Model Complexity:

Experiment with deeper networks or try using RNNs or LSTMs to handle the temporal dependencies in the data.

Consider integrating advanced techniques such as dueling DQN or prioritized experience replay for better learning performance.

- Exploration Strategy:

Implement epsilon annealing or use Boltzmann exploration to balance exploration and exploitation more effectively.

- Model Evaluation:

Conduct a more thorough evaluation of the model by testing on a variety of scenarios (including edge cases) to assess the generalization capability of the trained agent.

Conclusion

This project explored the application of deep reinforcement learning in energy management with a focus on battery charge and discharge decisions based on inverter power data. Although the agent failed to achieve positive rewards within the training period, we gained valuable insights into the model's limitations and areas for improvement. By addressing the reward function, model complexity, and training duration, we can enhance the agent’s performance and make it more suitable for real-world energy management tasks.
